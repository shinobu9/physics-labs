```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sec10 = pd.read_excel("liz10_1.xlsx") # 10 sec
sec40 = pd.read_excel("liz40_1.xlsx") # 40 sec
#sec10, sec40
```

```python
numsec10 = sec10.to_numpy()
numsec40 = sec40.to_numpy()
#numsec10, numsec40
```

```python
plt.bar(numsec10[:, 0], numsec10[:, 1], width = 1)
plt.bar(numsec40[:, 0], numsec40[:, 1], width = 1)
plt.show()
```

```python
from math import ceil
for i in range(len(numsec40)):
    numsec40[i][0] = numsec40[i][0]//4
```

```python
import seaborn as sns
N = 1000
data1 = []

for (a,b) in numsec10:
    tmp_list = [a for i in range(int(N*b))]
    data1 += tmp_list

data2 = []
for (a,b) in numsec40:
    tmp_list = [a for i in range(int(N*b))]
    data2 += tmp_list
```

```python
fig, ax = plt.subplots()

ax.bar(numsec10[:, 0], numsec10[:, 1], width = 1, color = 'grey')
#ax.bar(numsec40[:, 0], numsec40[:, 1], width = 1, color = 'orange')

appr = sns.kdeplot(data=data1, color = 'black')
#appr = sns.kdeplot(data=data2, color = 'red') # 2 

ax.set_xlabel('n')
ax.set_ylabel('w')
ax.legend(['10 ', '40 '], prop={'size': 20})

fig.set_figwidth(20)    #   Figure
fig.set_figheight(10)    #   Figure

import matplotlib.ticker as ticker
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.01))
```
